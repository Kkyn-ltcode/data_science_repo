```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = 'top50.csv'
top50_data = pd.read_csv(filepath, encoding='ISO-8859-1')
top50_data.head(5)
#show first 5 rows of the dataset
```

```python
#We need to remove 'Unnamed' column because it's unnecessary.
#Check the columns name
top50_data.columns
```

```python
#Remove column
top50_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
top50_data.head()
#now it's look better
```

```python
#check if dataframe has any missing value
top50_data.info()
```

```python
#Heatmap shows the correlation between each column in the dataset
plt.figure(figsize=(12,10))
sns.heatmap(data=top50_data.corr(), center=0.3, annot=True)
#corr() function collects all numerical columns in the top50_data and return a 
#correlation dataframe
plt.title('Correlation', size=15)
```

```python
top50_data.describe()
```

```python
plt.figure(figsize=(12,10))
top15 = top50_data.sort_values(by='Popularity',ascending=False).head(15)
sns.barplot(data=top15, orient='h', x='Popularity', y='Track.Name')
plt.ylabel('')
plt.xlabel('Popularity', size=15)
plt.title('Top 15 Most Popular Songs on Spotify in 2019', size=15)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
popular_genre = top50_data['Genre'].value_counts()
sns.barplot(y=popular_genre.index, x=popular_genre.values, orient='h')
plt.xlabel('Number of Songs', size=15)
plt.title('Most Popular Genre on Spotify in 2019', size=15)
```

```python
top10_artist = top50_data.groupby('Artist.Name')['Track.Name'].apply(lambda x: x.count()).sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.set_style('whitegrid')
sns.barplot(y=top10_artist.index, x=top10_artist.values, orient='h')
plt.ylabel('')
plt.xlabel('Number of Songs', size=15)
plt.title('Top 10 Artists Had Most Popular Song on Spotify in 2019', size=15)
```

```python
#Pair Plot shows relationship between a pair of columns
#from 3 columns: Popularity, Energy, Danceability
sns.set_style('darkgrid')
sub_data = top50_data.loc[:, ['Popularity', 'Danceability', 'Energy']]
sns.pairplot(data=sub_data, kind='kde', height=2.5)
```
