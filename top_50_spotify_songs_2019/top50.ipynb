```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
top50_data = pd.read_csv('top50.csv')
top50_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
top50_data.head()
```

```python
plt.figure(figsize=(12,10))
top15 = top50_data.sort_values(by='Popularity', ascending=False).head(15)
sns.barplot(data=top15, x='Popularity', y='Track.Name', orient='h')
plt.ylabel('')
plt.xlabel('Popularity', size=15)
plt.title('Top 15 Most Popular Songs on Spotify 2019', size=15)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
popular_genre = top50_data['Genre'].value_counts()
sns.barplot(y=popular_genre.index, x=popular_genre.values, orient='h')
plt.xlabel('Number of Songs', size=15)
plt.title('Most Popular Genre on Spotify in 2019', size=15)
```

```python
top10_artist = top50_data.groupby('Artist.Name')['Track.Name'].apply(lambda x: x.count()).sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 8))
sns.set_style('whitegrid')
sns.barplot(y=top10_artist.index, x=top10_artist.values, orient='h')
plt.xlabel('Number of Songs')
plt.title('Top 10 Artists Has Most Popular Song on Spotify in 2019')
```
